## Chapter 7 - Grouping Vs CrossTab Vs Pivot Table

```python
import pandas as pd
```

```python
insurance_data = pd.read_csv('insurance.csv')
insurance_data
```

```python
insurance_data.isna().sum()
```

```python
insurance_data.dtypes
```

```python
insurance_data.groupby()
```

```python
pd.crosstab(index = insurance_data['smoker'],columns = insurance_data['region'],margins=True)
```

### 1. Get the average insurance charge based on different regions

```python
insurance_data.groupby(by='region')['charges'].mean().sort_values(ascending = False)
```

### 2. Get the average insurance charge based on different regions and gender.

```python
insurance_data.groupby(by=['region','sex','smoker'])['charges'].mean().sort_values(ascending = False)
```

```python
pd.pivot_table(data = insurance_data,values='charges',index='sex',columns='region',aggfunc='mean').round()
```

```python
pd.pivot_table(data = insurance_data,values='charges',index='sex',columns=['region','smoker'],aggfunc='mean').round()
```
